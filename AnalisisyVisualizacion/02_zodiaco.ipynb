## Ejercicio 3:

**¿Cuál es la probabilidad de que entre “n” personas dos tengan el mismo signo?**

 $$
    p=P(A)=1-P(A^c)= \left\{\begin{array}{lr}
        1 & \text{si }  n> 12\\
        1-\frac{ 12!}{ (12-n)!\ 12^n} & \text{si } n\leq 12
        \end{array}\right.
 $$ 
 
 Depende de n, de la cantidad de personas

```python
def prob_coinc_signo(n):
    resultados = []
    if n>12:
        p = 1
    else: 
        aux = 1
        for k in range(n):
            aux = aux*(12-k)/12
            resultados.append(1-aux)
        p = 1-aux
    return resultados
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

## Elegir n: cantidad de personas (en una reunión)
n = 12

datos = pd.DataFrame({ 'mismosigno' : prob_coinc_signo(n) })

datos.plot(title='Dos personas con el mismo signo', color='r', figsize=(8,6))
plt.xlabel("asistentes")
plt.axvline(7)
plt.show()

# print('La probabilidad encontrar coincidencia de signo entre ',
#       n, ' personas es: ', prob_coinc_signo(n))
```

## Ejercicio 4

**¿Cuál es la probabilidad de que entre “n” personas dos hayan nacido el mismo día?**

 $$
    p=P(A)=1-P(A^c)= \left\{\begin{array}{lr}
        1 & \text{si }  n> 365\\
        1-\frac{ 365!}{ (365-n)!\ 365^n} & \text{si } n\leq 365
        \end{array}\right.
 $$ 
 
 Depende de n, de la cantidad de personas

Supongamos que el grupo sólo tiene una persona, en ese caso, hay una probabilidad del 100% que esta persona no comparte un cumpleaños puesto que no hay nadie más quien compartir. Pero ahora podemos añadir una segunda persona al grupo. ¿Cuáles son las posibilidades de que tenga un cumpleaños diferente de esa persona? De hecho esto es bastante fácil, hay 364 otros días en el año, así que las posibilidades son 364/365. ¿Qué tal si agregamos una tercera persona al grupo? Ahora hay 363/365 días. Para obtener la probabilidad general de que no hay cumpleaños compartidos simplemente multiplicamos las probabilidades individuales. Si utilizamos este procedimiento, con la ayuda de Python podemos calcular fácilmente las probabilidades de un cumpleaños compartido en un grupo de 50 personas.


```python
def prob_coinc_cumple(n):
    resultados = []
    if n > 365:
        p = 1
    else: 
        aux = 1
        for k in range(n):
            aux = aux*(365-k)/365            
            resultados.append(1-aux)
        p = 1-aux
    return resultados
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# %matplotlib inline

## Elegir n: cantidad de personas 
n = 70
df = pd.DataFrame({ 'coincidencias' : prob_coinc_cumple(n)})
# print('La probabilidad encontrar coincidencia de cumpleaños entre ',
#       n, ' personas es: ', prob_coinc_cumple(n))

df.plot(title='Posibilidad de que hayan nacido el mismo dia',color='r',figsize=(8, 6))
plt.xlabel("Numero de Invitados")
plt.show()
```
